```python
import pandas as pd
```

```python
data = pd.read_csv('C:\ArchTech_Internship\Task_01_email_Spam_detection_sys\spam_mail_classifier.csv')
```

```python
data.head()
```

```python
data.rename(columns={'email_text': 'input'}, inplace=True)
data.rename(columns={'label': 'output'}, inplace=True)
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.duplicated().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
# Encode labels
le = LabelEncoder()
y = le.fit_transform(data['output'])
```

```python
pd.DataFrame(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(data['input'],y,test_size=0.2,random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# 2️⃣ Fit vectorizer ONLY on training data
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)

# 3️⃣ Transform test data
X_test = vectorizer.transform(X_test)
```

```python
pd.DataFrame(X_train)
```

```python
pd.DataFrame(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
# Check accuracy on test set
y_pred = model.predict(X_test)
print("Model Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Get user input
email = input('Enter Your Email? ')

# Transform input using the same vectorizer
email_vector = vectorizer.transform([email])  # must use transform, not fit_transform

# Predict
res = model.predict(email_vector)

# Decode prediction
if res[0] == 0:
    print('Not a Spam Email')
else:
    print('Spam Email Detected!') 
```
